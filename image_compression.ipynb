```python
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread('/content/image.jpg')

cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from sklearn.cluster import KMeans
import numpy as np

def compress_image(image, k):
    pixels = image.reshape((-1, 3))

    kmeans = KMeans(n_clusters=k)
    kmeans.fit(pixels)

    centroids = kmeans.cluster_centers_
    labels = kmeans.labels_

    compressed_pixels = centroids[labels].astype(np.uint8)

    compressed_image = compressed_pixels.reshape(image.shape)

    return compressed_image
```

```python
k = 10

compressed_image = compress_image(image, k)
```

```python
cv2_imshow(compressed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
